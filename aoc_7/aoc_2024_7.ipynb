```python
import itertools
from itertools import product
```

```python
testinput = [
"190: 10 19",
"3267: 81 40 27",
"83: 17 5",
"156: 15 6",
"7290: 6 8 6 15",
"161011: 16 10 13",
"192: 17 8 14",
"21037: 9 7 18 13",
"292: 11 6 16 20"
]
```

```python
def makePer(n, concat=False):
    opt = ["+", "*"]
    if concat: opt += "||"
    return [list(c) for c in product(opt, repeat=n) ]

def eval_concat(reft):
    parts = reft.split("||")
    parts_eval = [str(eval(p)) for p in parts]
    out = ""
    for p in parts_eval: out+=p
    return int(out)
def test(ipt, pres=False, concat=False):
    split = ipt.split(": ")
    testvalue = int(split[0])
    rest=[int(v) for v in split[1].split(" ")]
    persN = len(rest)-1
    pers = makePer(persN, concat)
    if pres and not concat:
        minr = sum(rest)
        if testvalue < minr: return False, testvalue
        maxr = 1
        for v in rest: maxr *= v
        if testvalue > maxr: return False, testvalue
    for per in pers:
        ref = 0
        reft = ""
        for iv,v in enumerate(rest):
            if (not pres) and ref>testvalue: continue
            if iv == 0:
                ref = v
                reft = str(v)
            else:
                if per[iv-1] == "+": ref+=v
                elif per[iv-1] == "*": ref*=v
                else: ref = int(str(ref)+str(v))
                reft += per[iv-1] + str(v)
        if ((not pres) and ref == testvalue) or (pres and eval_concat(reft)==testvalue):
            print(testvalue, "=", reft)
            return True,testvalue
    return False, testvalue
```

```python
def count(ipt, concat=False, pres=False):
    c = 0
    for ii, i in enumerate(ipt):
        tc = test(i, concat=concat, pres=pres)
        if ii % max(1,int(len(ipt)/10)) == 0: print(ii, "/", len(ipt))
        if tc[0]: 
            c += tc[1]
    return c
```
