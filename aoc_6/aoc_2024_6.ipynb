```python
import numpy as np
```

```python
examplemap = [
"....#.....",
".........#",
"..........",
"..#.......",
".......#..",
"..........",
".#..^.....",
"........#.",
"#.........",
"......#..."
]
```

```python
# make a matrix out of it to work easier with indecies
def convertinput(ipt):
    ipt_conv = []
    for l in ipt:
        t = [c for c in l]
        ipt_conv.append(t)
    return np.array(ipt_conv)
```

```python
def findCurrentPosAndDir(ipt):
    direction = ""
    if "^" in ipt or "U" in ipt: direction="U"
    if "D" in ipt: direction="D"
    if "L" in ipt: direction="L"
    if "R" in ipt: direction="R"
    if len(direction) < 1: return False, 0, 0, ""
    pos = None
    if "^" in ipt:
        pos = np.where(ipt=="^")
    else: pos = np.where(ipt==direction)
    pos = [int(p) for p in pos]
    x = pos[1]
    y = pos[0]
    return True, x, y, direction

def updateDir(direction, x, y, ipt):
    if direction == "U":
        # left game or ahead is free
        if ((y-1)<0) or (ipt[y-1][x] != "#"):
            return "U"
            #try right
        elif ((x+1) >= (ipt.shape[1])) or (ipt[y][x+1] != "#"):
            return "R"
            #try down
        elif ((y+1) >= (ipt.shape[0])) or (ipt[y+1][x] != "#"):
            return "D"
            # must be left then
        return "L"
    elif direction == "R":
        # left game or ahead is free
        if ((x+1) >= (ipt.shape[1])) or (ipt[y][x+1] != "#"):
            return "R"
            #try down
        elif ((y+1) >= (ipt.shape[0])) or (ipt[y+1][x] != "#"):
            return "D"
            "try left"
        elif ((x-1)<0) or (ipt[y][x-1] != "#"):
            return "L"
            # must be up then
        return "U"
    elif direction == "D":
        # left game or ahead is free
        if ((y+1) >= (ipt.shape[0])) or (ipt[y+1][x] != "#"):
            return "D"
            "try left"
        elif ((x-1)<0) or (ipt[y][x-1] != "#"):
            return "L"
            # try up
        elif ((y-1)<0) or (ipt[y-1][x] != "#"):
            return "U"
            # must be right then
        return "R"
    elif direction == "L":
        # left game or ahead is free
        if ((x-1)<0) or (ipt[y][x-1] != "#"):
            return "L"
            # try uo
        elif ((y-1)<0) or (ipt[y-1][x] != "#"):
            return "U"
            # try right
        elif ((x+1) >= (ipt.shape[1])) or (ipt[y][x+1] != "#"):
            return "R"
            # must be down then
        return "D"
    
def updateMap(ipt):
    onmap, x, y , direction = findCurrentPosAndDir(ipt)
    if not onmap: return False, ipt
    # mark pos
    ipt[y][x] = "X"
    newdirection = updateDir(direction, x, y, ipt)
    newx = x
    newy = y
    if newdirection == "U": newy -= 1
    if newdirection == "R": newx += 1
    if newdirection == "D": newy += 1
    if newdirection == "L": newx -= 1
    if newx >=0 and newx < ipt.shape[1] and newy >=0 and newy < ipt.shape[0]:
        ipt[newy][newx] = newdirection
        return True, ipt
    else:
        return False, ipt
    
        
```

```python
def makestr(pos, direction):
    return "X="+str(pos[0])+"Y="+str(pos[1])+"dir="+direction
    
def testobstacle(poso, ipt):
    visisted=set()
    mp = ipt.copy()
    mp[poso[0], poso[1]]="#"
    omp, x,y, direction = findCurrentPosAndDir(mp)
    startpos = [x, y]
    visisted.add(makestr(startpos, direction))
    update = True
    while(update):
        update, mp = updateMap(mp)
        omp, x,y, direction = findCurrentPosAndDir(mp)
        newpos = [x, y]
        if makestr(newpos, direction) in visisted: 
            #print("found loop", newpos, visisted)
            return True # found loop
        else: visisted.add(makestr(newpos, direction))
    return False
```

```python
def getpotobs(ipt):
    # play one time
    mp = ipt.copy()
    update=True
    omp, x,y, direction = findCurrentPosAndDir(mp)
    startpos = [y, x]
    while(update):
        update, mp = updateMap(mp)
    posl = []
    #put obstacles where he would otherwise go
    posl1, posl2 = np.where(mp=="X")
    for ip,p in enumerate(posl1):
        if [int(p), int(posl2[ip])] != startpos:
            posl.append([int(p), int(posl2[ip])])
    return posl
```
